```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
scaled_data=scaler.fit_transform(data)

print("Original Data:\n", data)
print("Scaled Data:\n", scaled_data)
```

```python
# Visualization
plt.figure(figsize=(10, 5))

# Original data
plt.subplot(1, 2, 1)
plt.scatter(data['Age'], data['EstimatedSalary'], color='blue')
plt.title("Original Data")
plt.xlabel("Age")
plt.ylabel("Estimated Salary")

# Scaled data
plt.subplot(1, 2, 2)
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], color='orange')
plt.title("Scaled Data (MinMaxScaler)")
plt.xlabel("Scaled Age")
plt.ylabel("Scaled Estimated Salary")

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Combine original and scaled data for comparison
combined = pd.DataFrame(scaled_data, columns=['Age (scaled)', 'EstimatedSalary (scaled)'])
combined['Age'] = data['Age']
combined['EstimatedSalary'] = data['EstimatedSalary']

sns.pairplot(combined)
plt.suptitle("Pair Plot of Original vs Scaled Data", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(data['Age'], alpha=0.5, label='Age (Original)')
plt.hist(scaled_data[:, 0], alpha=0.5, label='Age (Scaled)')
plt.title("Age Distribution Before and After Scaling")
plt.legend()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(8, 5))

ax1.plot(data['EstimatedSalary'], label='EstimatedSalary (Original)', color='green')
ax1.set_ylabel('Original Salary')
ax1.set_title('Estimated Salary - Original vs Scaled')

ax2 = ax1.twinx()
ax2.plot(scaled_data[:, 1], label='EstimatedSalary (Scaled)', color='red')
ax2.set_ylabel('Scaled Salary (0 to 1)')

fig.legend(loc="upper left")
plt.show()

```
